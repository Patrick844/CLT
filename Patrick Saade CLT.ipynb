```python
from random import randrange
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy as scipy
    
```

```python
#Input Values
print("---- Your Values -----")
print("")
sample_size=int(input("Chose a the size of the sample : "))
nbr_sample=int(input("Chose the number of sample you need : "))
nbr_toss=int(input("Chose the number of toss you need : "))
nbr_outcomes=int(input("Chose the nbr of outcomes (ex 2 for toss of a coin) : "))

print("")
print("-------CLT ------- ")

#Call function
l2=[]
l2=CLT(sample_size,nbr_sample,nbr_toss,nbr_outcomes)

#Histogram
_,bins,_=plt.hist(l2, 10, density=1, stacked=1, alpha=0.5)

#Bell Curve
mu, sigma = scipy.stats.norm.fit(l2)
best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)

print("Mean = ", round(mu,3))
print("Standard deviation = ", round(sigma,3))

#Plot
plt.plot(bins, best_fit_line)
plt.title("Central Limit Theorem")
plt.legend(["Bell Curve","Histogram"])
plt.xlabel("Mean Values")

```

```python
def CLT(sample_size,nbr_sample,nbr_toss,nbr_outcomes):
    
    l1=[]
    l2=[]
    
    if sample_size >nbr_toss:
        raise ValueError('Sample Size cant be bigger then the number of experiment')
        
    l1=[ randrange(nbr_outcomes) for i in range(nbr_toss)]
    
    for i in range(nbr_sample):
        l_sample=random.sample(l1,sample_size)
        sample_sum=0

        for i in range(len(l_sample)):
            sample_sum+=l_sample[i]

        l2.append(sample_sum/sample_size)

    return l2
```
